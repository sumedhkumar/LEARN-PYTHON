### Higher Order Function

```python
# - One function is passed as a parameter to another function
# Eg. Map, Filter
```

```python
# Map(function name, iterable)

# Filter(Function name, Iterable)

# Map & filter both can work on condition but only map works on operations, filter does not! 

# If it is a condition then map will return TRUE & FALSE
# While Filter will return the actual values, 
```

```python
x1 = lambda x : x**2

x1(5)
```

```python
list1 = ["Jan", "Feb", "March", "Apr", "May"]
```

```python
x= lambda x: len(x)==3

x("jan")
```

```python
tuple(map(lambda x: len(x)==3, list1))
```

```python
tuple(filter(lambda x: len(x)==3, list1))
```

```python
list2 = [2, 3, 4, 5, 6, 7]
```

```python
list(map(lambda x : x**2, list2))
```

```python
list(filter(lambda x : x**2, list2))
```

```python
list(filter(lambda x : x**2==4, list2))
```

```python
list(filter(lambda x : x%2==0, [1, 2, 3, 4, 5 ,6]))
```

```python
list(filter(lambda x : x%2!=0, [1, 2, 3, 4, 5 ,6]))
```

```python
list(map(lambda x : x%2!=0, [1, 2, 3, 4, 5 ,6]))
```

```python
l1 = ['Sunday','Monday','Tuesday','Wednesday','thrusday','friday','Saturday']

list(map(lambda x:x.istitle(), l1))
```

```python
list(filter(lambda x:x.istitle(), l1))
```

```python
lst2 = [2, 3, 5]
list(map(lambda a : a+2, lst2))
```

```python
list(filter(lambda a : a+2, lst2))
```
