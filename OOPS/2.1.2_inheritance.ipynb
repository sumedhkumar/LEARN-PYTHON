### Inheritance

```python
# It allows a child class to inherit from parent classes

# Types of inheritance:
# 1. Single Inheritance: A child class inherits from one parent class.
# 2. Multiple Inheritance: A child class inherits from multiple parent classes.
# 3. Multilevel Inheritance: A child class inherits from a parent class, which in turn inherits from another parent class.
# 4. Hierarchical Inheritance: Multiple child classes inherit from a single parent class.
```

```python
class parent():
    print("this is parent class")
    
class child(parent):
    print("this is child class")
```

```python
class parent():
    def __init__(self, name):
        self.name = name
        # print("parent constructor")

    def parent_display(self):
        print("I am cricket player")
        
class child(parent):
    def __init__(self, name, age):
        # super().__init__(name)
        # parent.__init__(self, name)  # Alternative way to call parent constructor
        self.name = name  # Not needed if using super()
        self.age = age
        # print("child constructor")

    def child_display(self):
        print("I am football player")
        print("Name:", self.name)
        print("Age:", self.age)
```

```python
P = parent("Sachin")

P.parent_display()

C = child("Rohit", 18)
C.child_display()

C.parent_display()  # Accessing parent class method from child class instance

P.child_display()  # This will raise an error since parent class doesn't have child_display method
```

```python
class name():
    def __init__(self, name):
        self.name = name
        
    def student_name(self):
        print("Student name is:", self.name)
    
class student_info(name):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def student_info(self):
        print("Student name is:", self.name)
        print("Student age is:", self.age)
        
        
        
N = name("Ravi")
N.student_name()

S = student_info("Kavi", 20)
S.student_info()
```

```python
S.student_name()  # Accessing parent class method from child class instance
```

### 2. Multilevel Inheritance: A child class inherits from a parent class, which in turn inherits from another parent class.

```python
class grandparent():
    pass
class parent(grandparent):
    pass
class child(parent):
    pass
```

```python
class grandparent():
    def __init__(self, name):
        self.name = name
        print("Grandparent constructor")

    def grandparent_display(self):
        print("I am a grandparent")

class parent(grandparent):
    def __init__(self, name, age):
        self.name = name  # instance variable
        self.age = age
        print("Parent constructor")

    def parent_display(self):
        print("I am a parent")
        print("Name:", self.name)
        print("Age:", self.age)

class child(parent):
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
        print("Child constructor")

    def child_display(self):
        print("I am a child")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Grade:", self.grade)
```

```python
gp = grandparent("John")
gp.grandparent_display()
```

```python
# paretnt

p = parent("shahrukh", 40)
p.parent_display()
print("-------------------")
p.grandparent_display()  # Accessing grandparent class method from parent class instance
```

```python
# child
c = child("Ravi", 20, "A")
c.child_display()

print("-------------------")
c.parent_display()  # Accessing parent class method from child class instance
print("-------------------")
c.grandparent_display()  # Accessing grandparent class method from child class instance
# Multiple Inheritance
```

### 4. Hierarchical Inheritance: Multiple child classes inherit from a single parent class.


```python
# 4. Hierarchical Inheritance: Multiple child classes inherit from a single parent class.

class parent():
    def __init__(self, name):
        self.name = name
        print("Parent constructor")

    def parent_display(self):
        print("I am a parent")
        print("Name:", self.name)
class child1(parent):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print("Child1 constructor")

    def child1_display(self):
        print("I am Child1")
        print("Name:", self.name)
        print("Age:", self.age)
class child2(parent):
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        print("Child2 constructor")

    def child2_display(self):
        print("I am Child2")
        print("Name:", self.name)
        print("Grade:", self.grade)
class child3(parent):
    def __init__(self, name, city):
        self.name = name
        self.city = city
        print("Child3 constructor")

    def child3_display(self):
        print("I am Child3")
        print("Name:", self.name)
        print("City:", self.city)
```

```python
p = parent("John")
p.parent_display()

print("-------------------")

c1 = child1("Ravi", 20)
c1.child1_display()
c1.parent_display()  # Accessing parent class method from child class instance

print("-------------------")

c2 = child2("Kavi", "A")
c2.child2_display()
c2.parent_display()  # Accessing parent class method from child class instance
print("-------------------")
c3 = child3("Ravi", "Delhi")
c3.child3_display()
c3.parent_display()  # Accessing parent class method from child class instance
print("-------------------")
```

```python
## simple inheritance

### parent ---> child (parent + child)

## Multilevel inheritance

### grandparent (grandparent) ---> parent (grandparent + parent) ---> child (grandparent + parent + child ) ---> grandchild (grandparent + parent + child + grandchild)

## Multiple inheritance

### parent1 (parent1) + parent2 (parent2) ---> child (parent1 + parent2 + child)

# Hierarchical inheritance
# parent (parent) ---> child1 (parent + child1)
# parent (parent) ---> child2 (parent + child2)
# parent (parent) ---> child3 (parent + child3)
```
