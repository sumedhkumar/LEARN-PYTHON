```python
class mathematics:
    
    def addition(self, a=1, b=2):
        print("addition of numbers are ", a+b)
        
    def subscract(self, i, j):
        print("substraction of numbers are ", i - j)
        
    def multiplication(self, x, y):
        print("multiplication of numbers are ", x * y)
```

```python
class simplemaths:             #defining a class

    final=333                  #Class Variable/Static Variable

    def numbers(self,a,b,c):   #defining method by taking in arguemets
        self.a = a             #Instance variable created
        self.b = b
        self.c = c
        print('numbers are ', a,b,c)

    def sumation(self,x,y):
        print('sumation is ',x+y)

    def avg(self):
        print('avg is',(self.a+self.b+self.c)/3,self.final)

my_maths = simplemaths()             #defining the object for the class
my_maths.numbers(10,20,30)           #calling a method of the class using object
my_maths.sumation(100,200)
my_maths.avg()
```

```python
class mathematics: # defining a class
    global_var = 1000  # Class variable
    
    def addition(self, a=1, b=2): # Method with default arguments
        self.a = a   # Instance variable
        self.b = b
        print("addition of numbers are ", a+b, self.global_var)
        
    def subscract(self, i, j): # Method with parameters
        self.i = i
        self.j = j
        print("substraction of numbers are ", i - j)
        
    def multiplication(self): # Method without parameters
        print("multiplication of numbers are ", self.a * self.b * self.global_var)
```

```python
maths = mathematics()
maths.addition(10, 20)
maths.subscract(20, 10)
maths.multiplication()
```

##### Class Constructor

```python
# __init__ is a constructor in python. 
# It is used to initialize the instance variables of the class. 
# It is called when an object of the class is created. 
# It can take arguments and can also have default values.
```

```python
class car:
    def engine(self, engine_type):
        print("Engine is Created", engine_type)
        
car_obj = car() # initializing the object of the class
car_obj.engine("V8") # calling the method of the class using object
```

```python
class Car:
    def __init__(self, engine_type): # Constructor
        self.engine_type = engine_type # Instance variable
        
    def engine(self):
        print("Engine is Created", self.engine_type)
        
car_obj = Car("V6")
car_obj.engine() # calling the method of the class using object
```

```python
class College():
    def __init__(self, var, var2):
        self.var = var
        self.var2 = var2
        print("Constructor is called")
        
    def display(self):
        print(f"Display method is called {self.var} , {self.var2}")
        
    def __del__(self): # Destructor
        print("Destructor is called")
        
college_obj = College("Amit", "Jha") # Creating an object of the class
college_obj.display() # Calling the method of the class using object
```

```python
class college:
    
    def __init__(self, name, location): # Constructor
        self.name = name
        self.location = location
        
    def display(self):
        print("College Name: ", self.name)
        print("College Location: ", self.location)
```

```python
clg = college("MIT", "Pune") # Creating an object of the class
clg.display() # Calling the display method
```

```python
class stocks():
    def __init__(self, stock_name, stock_price): # Constructor
        self.stock_name = stock_name
        self.stock_price = stock_price
    
    def display(self):  #Utility method / functions/ Methods 
        print("Stock Name: ", self.stock_name)
        print("Stock Price: ", self.stock_price)
    
s1 = stocks("Reliance", 2500) # Creating an object of the class
s2 = stocks("TCS", 3500) # Creating an object of the class
s3 = stocks("HDFC", 1500) # Creating an object of the class
s4 = stocks("ICICI", 2000) # Creating an object of the class
s5 = stocks("HDFC", 1500) # Creating an object of the class
```

```python
s1.display() # Calling the display method of the class using object
```

```python
s2.display()
```

```python
s5.display()
```

```python
stk = stocks("ICICI", 2000) # Creating an object of the class
stk.display() # Calling the display method
```

```python
stk.stock_name = "HDFC" # Changing the stock name
stk.stock_price = 1500 # Changing the stock price
stk.display() # Calling the display method
```

```python
# types of defining variables in python
# 1. Instance variable: It is defined inside the constructor and is unique to each object of the class.
# 2. Class variable: It is defined inside the class and is shared by all objects of the class.
# 3. Static variable: It is defined inside the class and is shared by all objects of the class. It is not defined inside the constructor.
# 4. Global variable: It is defined outside the class and is accessible from anywhere in the program.
# 5. Local variable: It is defined inside a method and is accessible only within that method.
```

```python
class career:
    def __init__(self, stream, location):
        self.stream = stream
        self.location = location
        
    def display(self):
        print("Career Stream: ", self.stream)
        print("Career Location: ", self.location)
```

```python
c1 = career("Doctor", "Pune") 
c2 = career("Engineer", "Mumbai") 
c3 = career("Teacher", "Delhi")
c4 = career("Scientist", "Bangalore") 
```
