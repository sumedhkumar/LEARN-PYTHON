## Inheritance in Python

```python
# 4 types of Inheritance

1. Simple Inheritance
2. Multiple Inheritance
3. MultiLevel Inheritance
4. Hierarchical Inheritance
```

```python
# Simple Inheritance

class parent():
  def my_quality(self):
    print("I am a painter")

class child(parent):
  def child_qual(self):
    print("I am a dancer")


P = parent()

P.my_quality()

print("--------------")

C = child()

C.child_qual()
C.my_quality()
```

```python
class parent():

  def __init__(self, name, age):
    self.name = name
    self.age = age

    print("initialise")

  def parent_details(self):
    print(f"my name is {self.name} & my age is {self.age}")


class child(parent):

  def child_name(self):
    print("my parent name is ", self.name)

P = parent("Suresh", 62)

P.parent_details()
print("------------")

kid = child("Suresh", 62)

kid.child_name()
kid.parent_details()

```

### Multilevel Inheritance

```python
class grandparent():
  def gp_details(self):
    print("Dadaji")

class parent(grandparent):
  def parent_details(self):
    print("Papaji")

class child(parent):
  def child_details(self):
    print("Kid")

gp = grandparent()

gp.gp_details()

print("-----------")

parent = parent()
parent.parent_details()
parent.gp_details()

print("-----------")

kid = child()
kid.child_details()
kid.parent_details()
kid.gp_details()
```

### Multiple Inheritance

```python
class mother():
  def quality(self):
    print("I am a good cook")

class father():
  def quality_father(self):
    print("I am a superhero & I can sing")

class vella(father, mother):
  def my_quality(self):
    print("i am studying")


mom = mother()
mom.quality()

print("-----------")

papa = father()
papa.quality_father()

print("-----------")

kid = vella()

kid.my_quality()
kid.quality_father()
kid.quality()
```

# Hierarchical Inheritance:

```python
class parent():
  def __init__(self, name):
    self.name = name

    print("initialise")

  def my_income(self):
    print("bahut jada")

  def my_prop(self):
    print("200 acres")

class child1(parent):
  def my_income1(self):
    print("10 lacs")
    return super().my_income()

class child2(parent):
  def my_income2(self):
    print("12 lacs")
    return super().my_prop()


p = parent("ramesh")

p.my_income()

print("-----------")

kid1 = child1("Ramesh")
kid1.my_income1()
# kid1.my_income()

print("-----------")

kid2 = child2("Suresh")

kid2.my_income2()
```
