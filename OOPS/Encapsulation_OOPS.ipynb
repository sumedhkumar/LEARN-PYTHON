## Encapsulation in OOP

It is Used to hide or restrict the access of variables or methods

It is Used for protecting the data modification

```python
class City():

  name = "Pune" #----> Class variable

  def location(self):
    print("Location is: ", self.name)

place = City()

place.location()
```

```python
place.name = "Mumbai"  # Overwrite / Update

place.location()
```

```python
class City(): # Encapsulated

  __name = "Pune" #----> Class variable

  def location(self):
      print("Location is: ", self.__name)

place = City()

place.location()
```

```python
place.name = "Mumbai"  # Overwrite / Update

place.location()
```
