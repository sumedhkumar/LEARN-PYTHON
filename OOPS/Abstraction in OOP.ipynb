```python
from abc import ABC, abstractmethod

class ProjetBlueprint(ABC):

  @abstractmethod
  def new_login(self):
    pass

  @abstractmethod
  def process_docs(self):
    pass

  @abstractmethod
  def save_details(self):
    pass
```

```python
class ProjectCRM(ProjetBlueprint):

  def new_login(self):
    print("print the name and city")

  def process_docs(self):
    print("documents have been processed")

  def save_details(self):
    print("data has been stored on database")

proj = ProjectCRM()

proj.new_login()
proj.process_docs()
proj.save_details()
```
