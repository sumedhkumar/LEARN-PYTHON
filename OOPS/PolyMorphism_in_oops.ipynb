```python
import math

class circle1():

  def __init__(self, r1):
    self.r1 = r1

  def area(self):
    return math.pi * (self.r1**2)

class circle2():
  def __init__(self, r1):
    self.r1 = r1

  def area(self):
    return math.pi * (self.r1**2)


obj1 = circle1(2)

print(obj1.area())

obj2 = circle2(5)

obj2.area()
```

```python
import math

class circle1():

  def __init__(self, r1):
    self.r1 = r1

  def area(self): # redundant
    print("this is a parent class")
    return math.pi * (self.r1**2)

class circle2(circle1):
  def __init__(self, r1):
    self.r1 = r1

  def area(self):
    print("This is a child class")
    return math.pi * (self.r1**2)


obj1 = circle1(2)

print(obj1.area())
```

This only happens when we have same function name. In child and parent classes

If the name of functions from both classes are same

then the child class will overwrite the function from parent class

```python
child = circle2(4)

child.area()
```
