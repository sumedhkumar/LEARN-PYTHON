### How to modify a tuple

```python
tuple2 = (1, 2, 3, 4, 5) #immuatable 

list1 = list(tuple2) # convert tuple to list
print(list1) # mutable

list1[0] = 10 # mutable
print(list1) # mutable

tuple2 = tuple(list1) # convert list to tuple
print(tuple2) # immutable
```

```python
tuple2 = (1, 2, 3, 4, 5)

list1 = list(tuple2) # convert tuple to list
print(list1) # mutable
list1.pop(0) # mutable
print(list1) # mutable
tuple2 = tuple(list1) # convert list to tuple
print(tuple2) # immutable
```
