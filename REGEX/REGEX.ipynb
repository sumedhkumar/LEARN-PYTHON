## \w
Matches any letter, digit or underscore. Equivalent to [a-zA-Z0-9_].

```python
text = "(used someone has sent you a msg, i like aAI and data science __ Python 247)"

re.findall('\w{1,6}', text)
```

## \W
Matches anything other than a letter, digit or underscore. Equivalent to [^a-zA-Z0-9_]

```python
text = "(used someone has sent @you a msg, i like 6%aAI and data science __ Python 247)"

re.findall('\W{1,6}', text)
```

## Search

```python
text = "(used someone has sent @you a msg, i like 6%aAI and data science __ Python 247)"

match1 = re.search('\w{1,6}', text)
match1
```

## Match

It will check the match at zeroth position

```python
text = "used someone has sent @you a msg, i like 6%aAI and data science __ Python 247)"

match1 = re.match('\w{1,6}', text)
match1
```

```python
text = "@used someone has sent @you a msg, i like 6%aAI and data science __ Python 247)"

match1 = re.match('\w{1,6}', text)
match1
```

## \b

## \B (if there anything at the start it will consider)
Matches, without consuming any characters, immediately between a character matched by \w and a character not matched by \w (in either order). It cannot be used to separate non words from words.

```python
text = """_company name: _LogiRoute @Logistics
$email: [ops@logiroute.in](mailto:ops@logiroute.in)
contact: +91-9767123400
Description: Ops Manager specializing in supply chain digitization."""

re.findall(r"\b[a-zA-Z0-9]{1,15}", text)
```

```python
text = """_company name: _LogiRoute Logistics_
$email: [ops@logiroute.in](mailto:ops@logiroute.in)
contact: +91-9767123400
Description: Ops Manager specializing in supply chain digitization."""

re.findall(r"\b[a-zA-Z0-9]{1,15}\b", text)
```

## special charecter [.]

```python
# It is used for getting any one or more special/charecter
```

```python
text = """_company name: _LogiRoute Logistics_
$email: [ops@logiroute.in](mailto:ops@logiroute.in)
contact: +91-9767123400
Description: Ops Manager specializing in supply chain digitization."""

re.findall(".[a-zA-Z0-9]{1,15}.", text)
```

```python
text = """_company name: _LogiRoute Logistics_
$email: [ops@logiroute.in](mailto:ops@logiroute.in)
contact: +91-9767123400&
Description: Ops Manager specializing in supply chain digitization."""

re.findall(".\d{10}.", text)
```

## [+] charecter is used for getting the multiple same charecter match

```python
text = """my email: [sumedh.patil@@@@@@@@@@@@@email.com](mailto:sumedh.patil@email.com)
company name: InsightX AI Solutions
email: [contact@insightx.ai](mailto:contact@insightx.ai)
contact: +91-9876543210
Description: Lead Data Scientist focusing on NLP and enterprise AI tools."""

re.findall("\w{6}[.]\w{5}[@]+", text)
```

```python
text = """my email: [sumedh.patil@@@@@@@@@@@@@email.com](mailto:sumedh.patil@email.com)
company name: InsightX AI Solutions
email: [contact@insightx.ai](mailto:contact@insightx.ai)
contact: +91-9876543210#######
Description: Lead Data Scientist focusing on NLP and enterprise AI tools."""

re.findall("\d{10}.+", text)
```

```python
text = """Python is veryyyyyyyy easy""" #veryyyy

re.findall("\w{4}[y]+", text)
```

```python
re.findall('[a-z]{0,4}[y]+',text)
```

## Astric [*] - It return similar charecter multiple time

```python
text = """Python is veryyyyyyyy easy""" #veryyyy

re.findall("very*", text)
```

```python
data = "python pythoon pythhhhon pytttthon pythoooooon"

re.findall('pytho*n', data)
```

## {$} special charteter which is used to match only last match of the string

```python
text = """_company name: _LogiRoute Logistics_
$email: [ops@logiroute.in](mailto:ops@logiroutein
contact: +91-9767123400
Description: Ops Manager specializing in supply chain digitization_"""

re.findall('\w{7}$', text)
```
