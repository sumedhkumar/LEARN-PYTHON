### Comments In Python

Define: 
    Comments are used to explain the code
    To define comment use # 
    Prevent excution when running the code


```python
def is_prime(n):
    # Check if the number is less than or equal to 1 (not prime) <----- Comment added
    if n <= 1:
        return False

    # 2 is the only even prime number
    if n == 2:
        return True

    # Exclude other even numbers greater than 2 (not prime)
    if n % 2 == 0:
        return False

    # Check for factors from 3 to the square root of n (only odd numbers)
    for i in range(3, int(n**0.5) + 1, 2):
        # If a divisor is found, the number is not prime
        if n % i == 0:
            return False

    # If no divisors are found, the number is prime
    return True

# Example usage
# num = int(input("Enter a number: "))

# Check if the input number is prime and print the result
# if is_prime(num):
    # print(f"{num} is a prime number.")
# else:
    # print(f"{num} is not a prime number.")
```
