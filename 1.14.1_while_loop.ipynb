### While Loop

```python
# A while loop is used to execute a block of code repeatedly as long as a given condition is true.
# The syntax of a while loop is:
# while condition:
#     # code to be executed
#     # if condition is true
#     # code will be executed
#     # if condition is false   
#     # code will not be executed
#     # and the loop will terminate
#     # and the program will continue with the next statement after the loop
# Example 1: Basic while loop

count = 0
while count < 5:
    print("Count is:", count)
    count += 1  # Increment the count by 1
else:
    print("Count has reached 5, exiting loop.")    

    
# a = 4
# if a == 5:
#     print("a is 5")
# else:
#     print("Count is less than 5, continuing loop.")
```
