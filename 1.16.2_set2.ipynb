- Union

```python
# Union need seperate variable to store the change
# We can not add a tuple or a list inside a set
# It adds the whole set to the set
# It does not change the original set
# It returns a new set with the union of both sets
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

set1 = set1.union(set2)
print(set1)  # {1, 2, 3, 4, 5, 6, 7, 8}
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = (1)

set1 = set1.union(set2)
print(set1) 
```

```python
set1 = [1, 2, 3, 4, 5]
set2 = {4}

set1 = set1.union(set2)
print(set1)
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {"sunday", "monday", "tuesday"}
set3 = {"wednesday", "thursday", "friday"}

set1 = set1.union(set2, {6, 7, 8}, set3)
set1
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {"sunday", "monday", "tuesday"}
set3 = {"wednesday", "thursday", "friday"}

set1.union(set2, {6, 7, 8}, set3)
```

- Intersection

```python
# Basically it gives away the common items from the sets
```

```python
cities = {"Pune", "mumbai", "hyderabad"}
cities2 = {"Pune", "mumbai", "delhi"}

cities.union(cities2)  
```

```python
cities = {"Pune", "mumbai", "hyderabad"}
cities2 = {"Pune", "mumbai", "delhi"}

cities.intersection(cities2)  
```

```python
cities
```

```python
cities1 = {"Pune", "mumbai", "hyderabad", "delhi"}
cities2 = {"Pune", "mumbai", "delhi", "banglore"}
cities3 = {"Pune", "mumbai", "delhi", "banglore"}

common_cities = cities1.intersection(cities2, cities3)
print(common_cities)
common_cities = cities2.intersection(cities3)
common_cities
```

- Difference

```python
# It shows the uncommon items from the set
```

```python
cities = {"Pune", "mumbai", "hyderabad"}
cities2 = {"Pune", "mumbai", "delhi"}

cities.difference(cities2)
cities2.difference(cities)
```

```python
cities1 = {"Pune", "mumbai", "hyderabad", "delhi"}
cities2 = {"Pune", "mumbai", "delhi", "banglore"}
cities3 = {"Pune", "mumbai", "delhi", "banglore"}

uncommon_cities = cities1.difference(cities2, cities3)
print(uncommon_cities)
```

```python
new_actors = {"ranbir", "ranveer", "salman", "shahrukh"}
old_actors = {"amitabh", "shahrukh", "salman", "ajay"}

new = new_actors.union(old_actors)
print(new)

common_actors= new_actors.intersection(old_actors)
print(common_actors)

uncommon_actors= new_actors.difference(old_actors)
print(uncommon_actors)
```

- isdisjoint 

```python
# When there are no items common between 2 sets, then it return TRUE otherwise FALSE
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {6, 7, 8, 9, 10}

set1.isdisjoint(set2)  # It returns True if there are no common items in both the sets
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {6, 7, 8, 9, 10, 1}

set1.isdisjoint(set2)
```

- IS-SUBSET

```python
# When all the items in a set are also present in another set then it returns TRUE
```

```python
set1 = {1, 2, 3, 4, 5}
set2 = {1, 2, 3}

set2.issubset(set1)  # It returns True if all the items in set2 are present in set1
```

```python
set1 = {1, 2, 4, 5}
set2 = {1, 2, 3}

set2.issubset(set1)  # It returns True if all the items in set2 are present in set1
```

- IS-SUPERSET

```python
set1 = {4, 7, 9, 10}
set2 = {4, 7, 9, 10, 11}

set2.issuperset(set1)  # It returns True if all the items in set1 are present in set2
```

```python
set1 = {4, 7, 9, 10, 99}
set2 = {4, 7, 9, 10, 11}

set2.issuperset(set1) 
```
