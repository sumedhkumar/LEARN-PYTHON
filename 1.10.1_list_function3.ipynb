## Sorting in List

#### Ascending Sorting

- sort or sorted

```python
marks = [67, 76, 87, 98, 54, 90]

marks.sort()

marks
```

```python
sorted_marks_list = sorted(marks)

sorted_marks_list
```

```python
marks = [67, 76, 87, 98, 54, 90]

for i in range(len(marks)):
    print("i loop", i)
    for j in range(i + 1, len(marks)):
        print("j loop", j)
        if marks[i] >= marks[j]:
            marks[i], marks[j] = marks[j], marks[i]  #Swapping
        print(f"when i is {i}, j is {j}, list is {marks}")
marks
```

```python
# Does this sorting works with strings (words)

metro_cities = ["&sfsk", "Umbre" , "mum", "pune", "nag", "AAPle", "blr", "chn", "del", "amd", "&&&afk"]

metro_cities.sort()

metro_cities

# Special char < Upper case < Smaller case 
```

#### Decending Sorting

```python
marks = [67, 76, 87, 98, 54, 90]
marks.sort()   #---> List is acendingly sorted 
marks[::-1]    # Reversed it 
```

```python
marks = [67, 76, 87, 98, 54, 90]
marks.sort(reverse=True)

marks 
```

```python
marks = [67, 76, 87, 98, 54, 90] 

x = sorted(marks, reverse=True)

x
```
