## Global Variable and Local Variable

```python
# Global Variable: A variable defined outside function and has score of entire notebook 
# Local Variable: A variable defined inside function and has scope for that perticular method only
```

```python
variable1 = 800  #------> Global Variable

# Method / Function 
def function1():
    variable1 
    variable2 = 755 # -----> Local Variable 
    return variable2, variable1

print(function1())
# print(variable1)
```
