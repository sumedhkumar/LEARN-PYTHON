## SET 

```python
set1 = {1, 2, 1.2, "hello", (1, 2), 1}
set1
```

```python
# Unordered data type
# Mutable data type
# No duplicate elements

# Mutable ---> List, Dictionary, Set
# Immutable ---> Tuple, String, Integer, Float, Boolean (T/F)

# Sets are enclosed within { } ---> Curly braces
# Items in set are comma seperated

# It does not follow the indexing, slicing, or other sequence-like behavior 
```

```python
T  = 1, 2, 3, 4, 5

print((T))

set_new = set(T)
set_new
```
