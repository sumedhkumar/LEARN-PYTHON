### Set Functions In Python

#### -> How to access items in set

```python
set1 = {1, 2, 3, 4, 5}

for i in set1:
    print(i)
```

```python
set2 = {"summer", "winter", "spring", "autumn"} # unordered data type

for i in set2:
    print(i)
```

#### Set functions : To delete/remove items from the set

```python
# Remove
# Discard
# Pop
# Clear
# Del
```

- remove

```python
set1 = {1, 4, 1, 3, "heppy", 2, 1.2, (1, 2), 1}
set1.remove("heppy")
set1
```

```python
set2 = {"salman", "khan", "hello", 1, 2, 3, 4, 5}
set2.remove("danish")
set2
```

- Discard

```python
set1 = {1, 4, 1, 3, "heppy", 2, 1.2, (1, 2), 1}
set1.discard("heppy")
set1
```

```python
set2 = {"salman", "khan", "hello", 1, 2, 3, 4, 5}
set2.discard("danish")
set2
```

- Pop

```python
# pop function randomly deletes any element from the set
```

```python
list1 = [1, 2, 3, 4, 5]
list1.pop(0)
list1
```

```python
set1 = {1, 4, 1, 3, "heppy", 2, 1.2, (1, 2), 1}
set1.pop()
set1
```

```python
set1 = { 4, 3, "heppy", 2, 1.2, (1, 2)}
set1.pop()
set1
```

```python
set4 = {5, 25, 6, 36, 7, 49, 8, 64, 9, 81}

set4.pop()
set4
```

- Clear

```python
# It removes all the items from the set
```

```python
set3 = {1, 2, 3, 4, 5}
set3.clear()
set3
```

```python
set5 = {"carry", "on", "my", "wayward", "son"}
clear(set5)
set5
```

```python
set5 = {"carry", "on", "my", "wayward", "son"}
set5.clear("my")
set5
```

- Del

```python
# delete function 
```

```python
set1 = {"Saniya Mirza", "Rohit Sharma", "Virat Kohli", "MS Dhoni", "Sachin Tendulkar"}
set1
del set1
set1
```

```python
set1 = {"Saniya Mirza", "Rohit Sharma", "Virat Kohli", "MS Dhoni", "Sachin Tendulkar"}
set1
del set1[0]
set1
```

- Update

```python
set8 = {"kartik aryan", "sara ali khan", "deepika padukone", "ranveer singh"}

set9 = {"akshay kumar", "katrina kaif", "salman khan", "priyanka chopra"}
set8.update(set9)  # for list extend works
set8
```

```python
set10 = {11, 22, 33, 44}

set10.update({(100, 200), (100, 200)})

set10
```
