### FROZEN SET

```python
# set:
#     mutable
#     unordered
#     duplicates are not allowed
    
# Frozen Set:
#     immutable
#     unordered
#     duplicates are not allowed
```

```python
set1 = {1, 2, 3, 4, 5}
print(set1)
print(type(set1))


frozen = frozenset(set1)
print(frozen)
print(type(frozen))
```
