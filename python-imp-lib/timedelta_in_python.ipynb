### Time Delta

```python
import datetime
```

```python
curr_time = datetime.date.today()

print(curr_time)
```

```python
# Finding a past date

days = curr_time + datetime.timedelta(days = -10)
days
```

```python
# Finding a Future date (20 days)

days = curr_time + datetime.timedelta(days = 20)
days
```

```python
# Finding a past date for last 21st day

days = curr_time + datetime.timedelta(days = -21)
days
```

```python
current_time = datetime.datetime.now()

print(current_time)

hours = current_time + datetime.timedelta(hours = -5, minutes = -45, seconds = 11)
hours
```

### Strftime()

```python
# str f - time  >>> string from time object
# it returns a string
```

```python
current_time = datetime.datetime.now()
current_time
```

```python
current_time.strftime('%Y') # year
```

```python
current_time.strftime('%y') # year
```

```python
current_time.strftime('%m') # month
```

```python
current_time.strftime('%M') # Minute
```

```python
current_time.strftime('%d') # day
```

```python
## List of all different strftime formats for python
```

```python
## Python `strftime` Format Codes

| Format Code | Meaning                                 | Example      |
|-------------|------------------------------------------|--------------|
| `%a`        | Abbreviated weekday name                 | `Sun`        |
| `%A`        | Full weekday name                        | `Sunday`     |
| `%w`        | Weekday as a number (0=Sunday, 6=Saturday) | `0`        |
| `%d`        | Day of the month (01–31)                 | `12`         |
| `%b`        | Abbreviated month name                   | `May`        |
| `%B`        | Full month name                          | `May`        |
| `%m`        | Month as a number (01–12)                | `05`         |
| `%y`        | Year without century (00–99)             | `25`         |
| `%Y`        | Year with century                        | `2025`       |
| `%H`        | Hour (24-hour clock) (00–23)             | `14`         |
| `%I`        | Hour (12-hour clock) (01–12)             | `02`         |
| `%p`        | AM or PM                                 | `PM`         |
| `%M`        | Minute (00–59)                           | `30`         |
| `%S`        | Second (00–59)                           | `45`         |
| `%f`        | Microsecond (000000–999999)              | `123456`     |
| `%z`        | UTC offset in ±HHMM format               | `+0530`      |
| `%Z`        | Time zone name                           | `IST`        |
| `%j`        | Day of the year (001–366)                | `132`        |
| `%U`        | Week number of the year (Sunday as first day) | `19`   |
| `%W`        | Week number of the year (Monday as first day) | `20`   |
| `%c`        | Locale’s appropriate date and time       | `Mon May 12 14:30:45 2025` |
| `%x`        | Locale’s appropriate date                | `05/12/25`   |
| `%X`        | Locale’s appropriate time                | `14:30:45`   |
| `%%`        | A literal `%` character                  | `%`          |

```
