- Remove 

```python
# Remove is used to remove the item from the list 
# By default it removes the first occurance from the list always
```

```python
list_new = [2, 3, 4, 6, "iOS", "Andriod", 128]
list_new.remove("iOS")
```

- POP ()

```python
# pop(index) <----Syntax

# If we do not pass the index then by default it remove the last item from the list (-1th index)
```

```python
list1 = ["Dhone", "Virat", "Shahrukh", "Kajol", "sehwag", "kriti"]
list1.pop(2)
list1
```

```python
# Remove - need to mention the exact item name 
# Pop - Need to mention the index of the item 
```

```python
list1.pop(2)
list1
```

```python
list1.pop()
list1
```

```python
list1.pop(7)
```
