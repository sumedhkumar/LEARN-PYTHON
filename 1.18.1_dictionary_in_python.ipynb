- ZIP Function

```python
# back working of ZIP function

marks = [90, 80, 60, 99, 87]
names = ["John", "Jane", "Mike", "Tom", "Jerry"]
dict1 = {}

for i, key in enumerate(names):
    # print(i, key)
    for j in range(len(marks)):
        # print(j, marks[j], marks[j][i])
        if i == j:
            dict1[key] = marks[j]
            
dict1
```

```python
marks = [90, 80, 60, 99, 87]
names = ["John", "Jane", "Mike", "Tom", "Jerry"]
```

```python
results = list(zip(names, marks))
print(results)
# convert list of tuples to dictionary
final_results= dict(results)
final_results
```

```python
results = dict(zip(names, marks))
results
```

- Deep copy

```python
from copy import deepcopy
```

```python
technologies = {"LLM": 20000, "AI": 50000, "ML": 40000, "DL": 50000}
print("technology---->", technologies)

copy_tech1 = deepcopy(technologies)
print("copy_tech1---->", copy_tech1)

copy_tech1["LLM"] = 50000
print("copy_tech1---->", copy_tech1)
```

```python
# nested dictinary

technologies2 = {"LLM": {"name": "LLM", "price": 20000, "rating": 4.5},
                  "AI": {"name": "AI", "price": 50000, "rating": 4.8},
                  "ML": {"name": "ML", "price": 40000, "rating": 4.2},
                  "DL": {"name": "DL", "price": 50000, "rating": 4.7}}
print("technology2---->", technologies2)
```

```python
tech_copy2 = deepcopy(technologies2)
print("tech_copy2---->", tech_copy2)

tech_copy2["LLM"]["price"] = 55000
print("tech_copy2---->", tech_copy2)

print("technology2---->", technologies2)
```

```python
# shallow copy                               # deep copy 

1 works well with simple dict                1. works well with simple dict

2. does not work well with nested dict       2. works well with nested dict
(if I make changes in copy item, 
 those changes reflects in parent dict)  
```
