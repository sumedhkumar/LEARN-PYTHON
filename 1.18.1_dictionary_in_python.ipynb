#### How to update the values using UPDATE function

```python
emp1 = {"name": "John", "age": 30, "city": "New York", "salary": 50000, "department": "IT"}

new_data = { "emp_id": 101, "emp_name": "John", "emp_salary": 50000, "emp_department": "IT"}

empl = {"name2": "max"}
emp1.update(new_data)
emp1
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'d': 4, 'e': 5, 'f': 6}
dict1.update(dict2)
dict1 
```

- Setdefault()

```python
emp1 = {"name": "John", "age": 30, "city": "New York", "salary": 50000, "department": "IT"}
emp1.setdefault("name", "John")
emp1
```

```python
emp1 = {"name": "John", "age": 30, "city": "New York", "salary": 50000, "department": "IT"}
emp1.setdefault("id", 101)
emp1
```

```python
emp1 = {"name": "John", "age": 30, "city": "New York", "salary": 50000, "department": "IT"}
emp1.setdefault("passbook")
emp1
```

### Remove items from dictionary

- pop 

```python
new_dict = {"jan": 1, "feb": 2, "mar": 3, "apr": 4, "may": 5, "jun": 6, "jul": 7, "aug": 8, "sep": 9, "oct": 10, "nov": 11, "dec": 12}
new_dict
```

```python
new_dict.pop("apr")
new_dict
```

```python
new_dict.pop()
new_dict
```
