- Isnumeric()

```python
# True or false
# It supports all numbers & unicodes & subscipts and superscipts & vulgar fractions
# It does not allow special chars
```

```python
# Vulgar fractions unicodes
```

```python
string = "2395237"
string.isnumeric()
```

```python
string = "2390u523"

string.isnumeric()
```

```python
string = "23759 203857"
string.isnumeric()
```

```python
"\u2159".isnumeric()
```

```python
# Differences:

string = "234354"
print(string.isdecimal())
print(string.isdigit())
print(string.isnumeric())
```

```python
# with unicodes
string = "\u00B2"
print(string.isdecimal()) # not really work with unicdes
print(string.isdigit())
print(string.isnumeric())
```

```python
string = "\u2159" #--->1/6
print(string.isdecimal())
print(string.isdigit())
print(string.isnumeric()) # this allows vulgur fractions
```
