### Shallow copy & Deep Copy

- Shallow copy

```python
# When we have nested lists 
# shallow copy does not work well
# If we make change in the original list, the same change will reflect in copied list also
```

```python
list1 = [122, 4343, 5454, 545]

copied_list = list1.copy()
print(list1)
copied_list
```

```python
list1[0] = 1000
```

```python
print(list1)
copied_list
```

```python
# Nested List

list_n = [[2, 3], [4, 5], [6, 7]]
n_copied_list = list_n.copy()
print(list_n)
n_copied_list
```

```python
list_n[0][0] = 1000
```

```python
print(list_n)
n_copied_list
```

- Deep Copy

```python
from copy import deepcopy, copy
```

```python
# Nested List

list_n = [[2, 3], [4, 5], [6, 7]]
n_copied_list = deepcopy(list_n)
print(list_n)
n_copied_list
```

```python
list_n[0][1] = 1000111
```

```python
print(list_n)
n_copied_list
```
