## Numeric Data Type

```python
# Numberic Data types: 
    # Interger = int --> Whole number that could be +ve or -ve or 0 (Ranging from -infinity to +infinity)
    # Float = float --> Decimal number (Ranging from -infinity to +infinity) 
    # Boolean = bool --> True or False
    # Complex = complex --> Number with real and imaginary part (ex: 5+3j)
    # Binary = bin --> Binary representation of a number (ex: 0b1010)
    # Hexadecimal = hex --> Hexadecimal representation of a number (ex: 0x12c)
    # Octal = oct --> Octal representation of a number (ex: 0o123)
    
```

```python
# Integers 

A = 10, -10, 0


A = 10
print(type(A))
print(A)
```

```python
# Float

A = 10.1, -10.1, 0.0

A = 10.1
print(type(A))
print(A)
```

```python
# Boolean 

A = True, False

A = True
print(type(A))
print(A)
```

```python
# Complex

A = 10 + 20j
print(type(A))
print(A)
```

## Round Function In Python 

```python
# Understand Round function in python 

number = 89.888654 # Percentage of Ram
# 89.88% of Ram is used

# Round to 2 decimal places
Ram_score = round(number, 2)
print(Ram_score , end= " % ")
```
