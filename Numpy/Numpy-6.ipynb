## Copy (shallow copy) & Deep copy

### Copy

```python
# when you use copy function, the 2D or 3D array wont get affected like list functions
# Copy will work the same as deep copy
```

```python
original = np.array([2, 3, 4, 5, 6, 7])
# print(original)

xerox = np.copy(original)
# print(xerox)

xerox[0]=1000

print(original,"------", xerox)
```

```python
original = np.array([[2, 3, 4], [5, 6, 7]])
# print(original)

xerox = np.copy(original)
# print(xerox)

xerox[0][0]=10

print(original,"\n\n", xerox)
```

```python
# Understanding for LIST FUNCTIONS

original = [[2, 3, 4], [5, 6, 7]]
# print(original)

xerox = original.copy()
# print(xerox)

xerox[0][0]=10

print(original,"\n\n", xerox) # not desirable
```

## deepcopy

```python
from copy import deepcopy
```

```python
original = np.array([[2, 3, 4], [5, 6, 7]])
# print(original)

xerox = deepcopy(original) #deepcopy == np.copy both work the same
# print(xerox)

xerox[0][0]=10

print(original,"\n\n", xerox)
```

## Use of np.where to replace the values that does not matches the given condition in array

```python
arr = np.array([1, 2, 2, 2, 4, 3, 7, 8, 9])
print(arr)

np.where(arr%2==1, arr, 88)
```

```python
arr = np.array([1, 2, 2, 2, 4, 3, 7, 8, 9])
print(arr)

np.where(arr==2, arr, 11)
```

```python
arr = np.array([1, 2, 2, 2, 4, 3, 7, 8, 9])
print(arr)

np.where(arr==1, arr, 99) #replace
```

```python
arr = np.array([1, 2, 2, 2, 4, 3, 7, 8, 9])
print(arr)
np.where(arr%2==1, arr, 77)
```

## arg

### Argmax --> returns the index of the maximum number in the array

```python
A = [2, 4, 5, 6, 7]

np.argmax(A)
```

```python
x = [324, 54, 55, 33, 11]
np.argmax(x)
```

### argmin --> return the index of minimum number in te array

```python
A = [2, 4, 5, 6, 7]

np.argmin(A)
```

### argsort --> it returns the indexes of the sorted array

```python
x = [324, 54, 55, 33, 11]
np.argsort(x)

# 11, 33, 54, 55, 324
```

### Size --> it will give you the total elements in the array

```python
x = [324, 54, 55, 33, 11]

np.size(x)
```

```python
x = np.array([[324, 54], [55, 33]]) # it works with homogenious array only

np.size(x)
```
