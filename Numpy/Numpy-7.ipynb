## Mathematical Operations on Array

```python
import numpy as np
```

```python
array = np.array([2, 3, 4, 5, 6, 7, 8, 9])
array
```

- Addition

```python
array + 10
```

- Substraction

- Multiplication

```python
array = np.array([2, 3, 4, 5, 6, 7, 8, 9])
array * 5
```

- Floor Division

```python
array = np.array([2, 3, 4, 5, 6, 7, 8, 9])
array // 2
```

```python
array = np.array([[2, 3], [7, 9]])
array // 2
```

- Remainder

```python
array = np.array([2, 3, 4, 5, 6, 7, 8, 9])
array % 2
```

```python
array = np.array([[2, 3], [7, 9]])
array % 2
```

```python
array1 = np.array([2, 3, 4, 5])
array2 = np.array([8, 9, 7, 5])

array1 % array2
```

```python
array1 = np.array([2, 3, 4, 5])
array2 = np.array([8, 9, 7, 5])

np.remainder(array1, array2)
```

### Matrix dot product
(dot --> multiplication)

```python
# for first column

(a1) first row * (a2) first column

(a1) second row * (a2) first column

# for second column

(a1) first row * (a2) second column

(a1) second row * (a2) second column
```

```python
arr1 = np.array([[3, 4],
                 [5, 5]])

arr2 = np.array([[5, 6],
                 [3, 4]])

3 * 5 + 4 * 3 --- 3 * 6 + 4 * 4
5 * 5 + 5 * 3 --- 5 * 6 + 5 * 4
arr1, arr2
```

```python
np.dot(arr1, arr2)
```

```python
arr1 = np.array([[3, 4],
                 [5, 5],
                 [0, 1]])

arr2 = np.array([[4, 6, 1],
                 [9, 0, 2]])

np.dot(arr1, arr2)
```

```python
arr1 = np.array([[3, 4],
                 [5, 5],
                 [0, 1]])

arr2 = np.array([[4, 6],
                 [5, 6]])


np.dot(arr1, arr2)
```

```python
arr1 = np.array([[3, 4],
                 [5, 5]])

arr2 = np.array([[4, 6],
                 [5, 6]])

arr3 = np.array([[0, 0],
                [0, 0]])

step1 = np.dot(arr1, arr2)

step2 = np.dot(step1, arr3)
```

```python
arr1 = np.array([[3, 4],
                 [5, 5]])

arr2 = np.array([[4, 6],
                 [5, 6]])

arr3 = np.array([[0, 0],
                [0, 0]])


np.dot(arr1, arr2, arr3)
```
