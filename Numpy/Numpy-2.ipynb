## Reshape

```python
array = np.array([2, 4, 5, 6, 7, 8, 9, 0 ,0 ,88, 99, 77])
array
```

```python
print(array.reshape(1, 12))
print(array.reshape(12, 1))
```

```python
array.reshape(3, 4)
```

```python
array.reshape(4, 3)
```

```python
array.reshape(2, 6)
```

```python
array.reshape(6, 2) # 6 lists (each list contains 2 elements)
```

```python
nums = np.array([2, 3, 4, 5])
nums
```

```python
nums.reshape(2, 2)
# nums.reshape(1, 4)
```

```python
# 3D examples
```

```python
array.reshape(2, 3, 2)
```

```python
array.reshape(3, 2, 2)
```

```python
array.reshape(2, 2, 3) #(rows, rows, columns)
```

```python
array.reshape(1, 2, 3) # because you had 12 elements in the array, and this is taking only 6 elements
```
