## How to generate Random Number

```python
import numpy as np
import random
```

```python
# There are multiple functions to create Random Numbers

random.random() #[0-1] use reshape functions to get 2D or 3D array
random.rand() #[0-1]
random.randint() #[low, high, size, dtype]
random.ranf() #[0-1] use reshape functions to get 2D or 3D array
random.randn() # (0,0) may contain negative values
```

### Random.rand()
- It will generate values between 0-1

```python
nums = np.random.rand(7)
nums
```

```python
nums = np.random.rand()
nums
```

```python
nums = np.random.rand(10)
nums
```

```python
nums = np.random.rand(4, 5)
nums
```

```python
nums = np.random.rand(5, 10)
nums
```

### random.random()

```python
random_nums = np.random.random(10)
random_nums
```

```python
random_nums = np.random.random(7)
random_nums
```

```python
random_nums = np.random.random(4).reshape(2, 2)
random_nums
```

```python
r_nums = np.random.random(12)
r_nums.reshape(3, 2, 2)
```

## Random.ranf()
- Values between 0-1

```python
np.random.ranf(7)
```

```python
np.random.ranf(12).reshape(6, 2)
```

## random.randint()

```python
np.random.randint(1, 20, 15)
```

```python
np.random.randint(1, 20, size = 15, dtype= 'I')
```

```python
np.random.randint(low=1, high=20, size=15, dtype='I')
```

## random.randn()
- it will generate the normally distributed numbers around (0, 0) cordinates

```python
np.random.randn(7)
```

```python
np.random.randn(5, 4)
```

## Sort()

```python
ram_num = np.random.randint(low=1, high=20, size=15, dtype='I')
ram_num
```

- ascending

```python
np.sort(ram_num)
```

```python
num = np.array(sorted(ram_num, reverse=False))
num
```

- descending

```python
np.sort(ram_num)[::-1]
```

```python
num = np.array(sorted(ram_num, reverse=True))
num
```

## Sorting of Matrix

```python
np.sort([2, 3, 5, 0], axis=-1) # rows
```

```python
array = np.array([1, 3, 44, 33, 9, 6, 0, 33, 4, 1, 55, 3]).reshape(4, 3)
array
```

- Sorting columns wise

```python
array = np.array([1, 3, 44, 33, 9, 6, 0, 33, 4, 1, 55, 3]).reshape(4, 3)
array
```

```python
np.sort(array, axis=1)
```

```python
ary = np.array([22, 2, 33, 3, 44, 4]).reshape(2, 3)
ary
```

```python
np.sort(ary, axis=1)
```

- Sorting raw wise

```python
array = np.array([1, 3, 44, 33, 9, 6, 0, 33, 4, 1, 55, 3]).reshape(4, 3)
array
```

```python
np.sort(array, axis=0)
```

```python
ary = np.array([22, 2, 33, 3, 44, 4]).reshape(2, 3)
ary
```

```python
np.sort(ary, axis=0)
```
