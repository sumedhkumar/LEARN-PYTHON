### linspace() --> Creates a linear space values

```python
dummy_arr = np.linspace(0, 20)
dummy_arr
```

```python
dummy_arr = np.linspace(0, 20, num=21, dtype="I")
dummy_arr
```

```python
dum_arr = np.linspace(0, 23 , 12)
dum_arr
```

```python
d_arr = np.linspace(1, 21 , 11)
d_arr
```

```python
d_arry = np.linspace(100, 1000, 10) # (start Num, End Num, Numbers of elements)
d_arry
```

```python
dict1 = {
    'amount': np.linspace(100, 600, 6, dtype = int),
    'interest': np.linspace(12.5, 3.5, 6),
    'month': np.linspace(6, 36, 6, dtype=int)
}
dict1
```

```python
df = pd.DataFrame(dict1)
df
```
