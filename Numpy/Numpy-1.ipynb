## ndim --> To check the dimensions of an array

```python
# single dimension --> [2, 4, 6, 8]
# double dimension --> [[2, 4, 6, 8],[2, 4, 6, 8]]
# three dimension --> [[[3, 5]]]
# four dimension ---> [[[[2, 3]]]]
# five dimension --> [[[[[2, 3]]]]]
```

```python
array = np.array([2, 4, 6, 8])
array
```

```python
array.ndim
```

```python
array = np.array([[2, 4, 6, 8],[2, 4, 6, 8]])

array.ndim
```

```python
array = np.array([[[3, 5], [6, 7]]])
array.ndim
```

```python
ary= np.array([[[2, 3], [5, 6]]])

ary.ndim
```

```python
array=np.array([[[[2, 3]]]])
array.ndim
```

```python
array=np.array([[[[[2, 3]]]]])
array.ndim
```

```python
array = np.array(2.3)
array.ndim
```

## ndmin --> converting lower dimensions array into higher dimensions array

```python
array = np.array(2.3, ndmin=3) # 0D --> 1D ---> 2D ....
print(array)
array.ndim
```

```python
array = np.array([[[2, 3], [4, 5]]], ndmin=1) # 3D array ---> 1D (we can not reduce higher dimension array into a lower dimension array)
print(array)
array.ndim
```

```python
num = np.array([[3, 4], [3, 4]], ndmin=3)
num.ndim
num
```
