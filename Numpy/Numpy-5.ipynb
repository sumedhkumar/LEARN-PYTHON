## Inserting item in array

```python
# insert - wheneven you want to add any specific new item (syntax: np.insert(array, index, value))
# append
# concatinate
```

```python
ary =np.array([1, 2, 3, 4, 5])

np.insert(ary, 2, 4000)
```

```python
np.insert(ary, len(ary), 5000) # wheneven you want to change the very last item. use len(ary)
```

```python
np.insert(ary, -1, 456)
```

## Looping on array

```python
list1 = [2, 3, 4, 5, 6]
for i in list1: #for loop & while loop
  print(i)
```

```python
aray1  = np.array([2,3, 4, 5, 5, 6 ,6, 7])
aray1
```

```python
for i in aray1:
  print(i)
```

```python
array2 = np.array([[2,3, 4, 5, 5,6 ,6, 7]])
array2
```

```python
for i in array2:
  print(i)
```

```python
array2 = np.array([[2,3, 4, 5, 5, 6,6, 7], [23, 34, 45], [44, 55, 66]]) # it wont work in multiple array items
for i in array2:
  print(i)
```

```python
array2 = np.array([2, 3, 4, 5, 5, 6, 6, 7])
array2
```

```python
for i in enumerate(array2):
  print(i)
```

```python
for index, value in enumerate(array2):
  print(index, "---", value)
```

### nditer()

```python
array2 = np.array([2, 3, 4, 5, 5, 6, 6, 7]).reshape(4, 2)
array2
```

```python
for i in np.nditer(array2):
  print(i)
```

```python
for i, v in np.ndenumerate(array2):
  print(i, "---", v)
```
