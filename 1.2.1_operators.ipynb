### Logical Operators

```python
# Logical Operators:
#     and  >> True if both the operands are true
#     or   >> True if either of the operands is true
#     not   >> True if operand is false
```

```python
a = 5
b = 6

if a == 5 and b == 6: # True and True = True
    print("Both are correct")
```

```python
a = 100000
b = 699
c = 27

if a == 100000 and b == 699 and c == 7: # True and True and True = True
    print("Loan Approved")
else:
    print("Loan Rejected")
```

```python
a = 5
b = 6

if a == 5 and b == 60: # True and False = False
    print("Both are correct")
else:
    print("either values are not correct")
```

```python
# Or operator

a = 5
b = 6

if a == 5 or b == 60: # True or False = True
    print("Either of the values are correct")
else:
    print("Both are not correct")
```

```python
# Or operator

a = 5
b = 6

if a == 5 or b == 6: # True or True = True
    print("Either of the values are correct")
else:
    print("Both are not correct")
```

```python
# Not Operator

a = 5
b = 6

if not a == 5: # False
    print("1st output: Value is not correct")

if not a == 6: # True
    print("2nd output: Value is not correct")
```

### Membership Operators

```python
# Membership Operators:
#     in
#     not in
```

```python
my_list = [1, 2, 3, 4, 5]

if 5 in my_list:
    print("5 is present in my list")
else:
    print("5 is not present in my list")
```

```python
my_list = [1, 2, 3, 4, 5]

if 6 not in my_list: # True
    print("6 is not present in my list")    
else:
    print("6 is present in my list")
```

### Comparison Operator

```python
# == (Equal to equal to)
# != (Not equal to)
# > (Greater than)
# < (Less than)
# >= (Greater than or equal to)
# <= (Less than or equal to)
```

```python
10 == 10 # True
10 != 10 # False
10 > 100 # False
10 < 100 # True
10 >= 10 # True
10 >= 100 # False
10 <= 10 # True
10 <= 100 # True
```

### Identity Operators

```python
# Identity Operators
#     is
#     is not
```

```python
a = 5
b = 5

if a is b:
    print("Both are same")
else:
    print("Both are not same")
```

```python
a = 5
b = 6

if a is not b:
    print("Both are not same")
else:
    print("Both are same")
```
