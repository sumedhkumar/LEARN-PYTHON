### APPEND Function

```python
# Append function is used to add any items in the list, 
# By default it adds the item at the end (-1th index)
# When multiple items to append, they will be added as a sublist
```

```python
lst = [1, 2, 3, 4, 5]

lst.append(77)
lst
```

```python
lst = [1, 2, 3, 4, 5]

lst.append([66, 88])
lst
```

```python
football = ["cristiano", "messi", "Neymar", "Zadden"]

football.append(["Virat", "Rohit", "dhoni"])

football
```

### EXTEND Function

```python
# Extend is used to add the items in list 
# by default it adds items at the -1th index (last index)
# When multiple items to be added, they will get added as a single singel items 
```

```python
football = ["cristiano", "messi", "Neymar", "Zadden"]

football.extend(["Virat", "Rohit", "dhoni"])

football
```

```python
# Q. Given a input, segrigate numbers into even and odds list

input = "128943984129"
# append function
# Operators % remainders (2 % 2=0)

even = []
odd = []

for i in input:
    if int(i) % 2 == 0: # TRUE
        even.append(int(i))
    else:
        odd.append(int(i))
        
print("even: ", even), print("odd", odd)
```

```python
input = [1, 2, 3, 5, 6, 7]
# append function
# Operators % remainders (2%2=0)

even = []
odd = []

for i in input:
    print(i)
    if i % 2 == 0: # TRUE=EVEN/FALSE=ODD
        even.append(i)
    else:
        odd.append(i)
        
even, odd
```

### Index function in List

```python
# It used to find the index of any item in the given list 

# syntax = list.index(item, occurance)
```

```python
Age = [10, 20, 34, 67, 80, 19, 23]

# Max function 

print(max(Age)) # it works only on integers 

Age.index(max(Age))
```

```python
Age.index(20)
```
