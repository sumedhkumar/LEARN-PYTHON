```python
# Even Odd Number 

list2 = [3, 2, 1, 5, 6, 7, 8, 10]

for value in list2: # checking numbers frm list
    # print(value)
    if value % 2 == 0: #True = The number is EVEN 
        print(value)
```

```python
# Even Odd Number 
list2 = [3, 2, 1, 5, 6, 7, 8, 10]

even = []
odd = []

for value in list2:                # checking numbers frm list
    print(value)
    if value % 2 == 0:             #True : The number is EVEN 
        even.append(value)         # to add any value into the list
    else: 
        odd.append(value)
        
even, odd
```

```python
# Seperate out int and strings from the list

list3 = [2, 3, 4, 5, 6, "Seven", 8, "Nine"]
number = []
word = []
for i in list3:
    # print(i)
    if type(i)==int:
        number.append(i)
    else:
        word.append(i)
    
number, word
```

### Nested Loop

```python
# for loop inside another loop
```

```python
list(range(5))
```

```python
for i in range(5):
    print("Outer Loop")
    print(i)
    
    for j in range(5):
        print("Inner Loop")
        print(i, j)
```

```python
for i in range(5):    # [0, 1, 2, 3, 4]
    print("print i", i)
    for j in range(5):    # [0, 1, 2, 3, 4]
        print("print j", j)
        for k in range(5):    # [0, 1, 2, 3, 4]
            print("print k", k)
```

```python
for i in "TATA":
    print("outter loop")
    
    for j in "BATA":
        print(i, j)
```

```python
for i in range(1): # [0]
    print("Outer Loop")
    # print(i)
    
    for j in range(2): # [0, 1]
        print("Inner Loop")
        print(i, j)
```
