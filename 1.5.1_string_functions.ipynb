- lstrip

```python
# lstrip ---> left strip (remove unnnesassry spaces) 
```

```python
country = "            india is my country, all indians are my brothers and sisters"
country = country.lstrip()
print(country)
```

```python
country = "            india is my country, all indians are my brothers and sisters"
country = country.lstrip()
print(country)
country.capitalize()
country.upper()
```

- rstrip 

#### It is right strip (remove unnnesary space)

```python
country = "india is my country, all indians are my brothers and sisters             "

country = country.rstrip()
country
```

- strip

```python
country = "         india is my country, all indians are my brothers and sisters             "
country.strip()
```

- Swapcase

```python
country = "         india is my country, all indians are my brothers and sisters             "
country = country.strip()
country = country.title()
country
```

```python
country.swapcase()
```
