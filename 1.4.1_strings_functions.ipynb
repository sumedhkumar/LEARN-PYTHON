### Indexing in String

Indexing is a way to access the item from the collection of items

1. Positive Indexing: 
Positive indexing starts from left to right . It starts from zero to infinity 
2. Negative Indexing:
Negative Indexing starts from right to left. It starts from -1 to negative infinity. 


```python
# lenght function

# len() --> Syntax

string = "Hello World"

len(string)
```

### String Slicing

```python
# String slicing 
# string[start index : end index : step size]
# The default of start index is 0
# The default of end index is length of string
# The default of step size is 1

# step size is positive --> Left to Right
# step size is negative --> Right to Left
```

```python
string = "Python"
string[0], string[-1]
```

```python
string[0:2]
```

```python
string[:2]
```

```python
string[0:6]
```

```python
len(string)
```

```python
string[0:len(string)]
```

```python
string[0:6] # ---> string[:]
```

```python
weature = "Winter is coming"
```

```python
weature[0:6]
```

```python
# weature[10:len(weature)]
weature[10:]
```

```python
# Step Size  ---> default step size is 1 
weature = "Winter is coming"
weature[0:15:1] # step size 1
weature[0:15:2] # step size 2
weature[0:15:3] # step size 3
```

### Negative Index slicing

```python
# Positive Index Slicing 
# Start index has to be always less than end index

# Negative IndeX Slicing
# Start index has to be always smaller than end index
```

```python
string = "I am learning Python"
```

```python
string[-6:]
```

```python
string[-15:-7]
```

```python
string[10:-1]
```

```python
# Q : Given A string reverse the string without using reverse function
string = "I like Python"
```

```python
string[::-1] # negative step size -1 
```
