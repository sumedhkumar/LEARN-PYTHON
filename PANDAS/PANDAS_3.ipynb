## Treatment

### drop na (delete the null column)

We should use this when a column has all the values as NaN

```python
df1 = df.dropna()
df1
```

```python
df
```

Fill the column with some data

```python
df.fillna("AI")
```

Fill na values by columns

```python
# calculate a mean

mean = df['Expense'].mean()
mean
```

```python
df['Expense'] = df['Expense'].fillna(mean)
df
```

```python
df['Expense'].median()
```

```python
df['Cashback'] = df['Cashback'].fillna(df['Cashback'].median())
df
```

```python
## Backward fill & Forward Fill
```

```python
df['Need & wants'] = df['Need & wants'].fillna(method='bfill')
df
```

```python
df['Need & wants'] = df['Need & wants'].fillna(method='ffill')
df
```
