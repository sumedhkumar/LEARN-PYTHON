### Subset
- Drop null values for specific to any column or rows

```python
df.dropna(subset=['Cashback'])
```

```python
df.dropna(subset=['Unnamed: 8'])
```

```python
df.dropna(subset=['card_name', 'month'])
```

```python
df.dropna(subset=['card_name', 'month'], how='any', axis=0)
# df + fun + fun2 + column names         + Kaise?   + kya delete karna hai?
```

```python
df.dropna(subset=['card_name', 'month'], how='all', axis=0)
```
