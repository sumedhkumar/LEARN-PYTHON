## Group by functions

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Jai', 'Anuj', 'Jai', 'Princi', 'Gaurav', 'Anuj', 'Princi', 'Abhi', 'Anuj'],
    'Age': [27, 24, 22, 32, 33, 29, 27, 32, 32],
    'Address': ['Nagpur', 'Pune', 'Kanpur', 'Allahabad', 'Jodhpur', 'Kanpur', 'Allahabad', 'Aligarh', 'Pune'],
    'Qualification': ['Msc', 'MA', 'MCA', 'PHD', 'BE', 'B.com', 'Msc', 'MA', 'B.com'],
    'Score': [24, 24, 35, 45, 53, 47, 50, 52, 60]
}

df = pd.DataFrame(data)
df
```

```python
# Q: Write a code to get the avg score for different degrees
df.groupby(['Qualification'])['Score'].agg(np.mean)
```

```python
df.info()
```

## Get_group

```python
### Groups

df.groupby('Name').groups
```

```python
df.groupby('Address').groups
```

```python
### ngroups --> it will return the number of unique values

df.groupby('Address').ngroups
```

## Size in groupby

```python
df.groupby('Name').size()
```

## Sort in groupby

```python
# group by is by default sorted
```

```python
df.groupby(['Score'], sort=True).mean('Age')
```

```python
df.groupby(['Score'], sort=False).mean('Age')
```

## Statistical Functions on groupby

```python
# Given a dataset, give avg age of students
```

```python
df.groupby(['Name'], sort=False).mean('Age') #avg/mean
```

```python
df.groupby(['Name'], sort=False).median('Age')
```

```python
df.groupby(['Name'], sort=False).median('Age').reset_index()
```

```python
df.groupby(['Name'], sort=False)['Age'].sum().reset_index()
```

```python
df.groupby(['Name'], sort=False)['Age'].min().reset_index()
```

```python
df.groupby(['Name'], sort=False)['Age'].max().reset_index()
```

```python
df.groupby(['Name'], sort=False)['Age'].count().reset_index()
```

## Agg functions in groupby

```python
df.groupby(['Name'])['Age'].agg(np.sum)
```

```python
df.groupby(['Name'])['Age'].agg([np.sum, np.mean, np.median, np.std])
```

## Aggregate function in groupby

```python
df.groupby(['Name', 'Qualification']).aggregate(np.sum)
```

```python
df.groupby(['Name', 'Qualification']).aggregate(np.sum).reset_index()
```

```python
def addition(var):
  return var + 2

addition(2)

df['Age'] = df['Age'].apply(addition)
df
```

```python
def uppercase(var):
  return var.upper()

uppercase("India")


df['Qualification'] = df['Qualification'].apply(uppercase)
df
```
