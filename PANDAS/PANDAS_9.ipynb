# Joining the DataFrames

## Append function

- Appending along the rows

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'ASCII_Num': np.arange(65,71),
                    'char': [chr(i) for i in range(65, 71)]})

df1
```

```python
df2 = pd.DataFrame({'ASCII_Num': np.arange(71,76),
                    'char': [chr(i) for i in range(71, 76)]})

df2
```

```python
df1.append(df2)
```

```python
display(df1, df2)
```

## Concat

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
df3 = pd.DataFrame({'ASCII_Num': np.arange(71,75),
                    'characters': [chr(i) for i in range(71, 75)]})

df3
```

```python
display(df1, df3)
```

```python
pd.concat([df1, df3], axis=0, ignore_index=True)
```

```python
df4 = pd.DataFrame({'ASCII_Numbers': np.arange(71,75),
                    'characters': [chr(i) for i in range(71, 75)]})

df4
```

```python
display(df1, df2, df3, df4)
```

```python
pd.concat([df1, df4], axis=0, ignore_index=True)
```

### Concat along the columns

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df3], axis=1)
```

```python
pd.concat([df1, df4], axis=1)
```

## Join parameters in concat

```python
pd.concat([df1, df4], axis=1, join='inner') # Keeps the common rows only
```

```python
pd.concat([df1, df4], axis=1, join='outer') # Keeps all the common rows
```

```python
pd.concat([df1, df3], axis=0, join='inner') # Keeps the common columns only
```

```python
pd.concat([df1, df3], axis=0, join='outer') # Keeps all the common columns
```
