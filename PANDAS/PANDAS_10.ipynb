# MERGE

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'ASCII_Num': np.arange(65,71),
                    'char': [chr(i) for i in range(65, 71)]})

df1
```

```python
df2 = pd.DataFrame({'ASCII_Num': np.arange(71,76),
                    'char': [chr(i) for i in range(71, 76)]})

df2
```

```python
df3 = pd.DataFrame({'ASCII_Num': np.arange(69,75),
                    'characters': [chr(i) for i in range(69, 75)]})

df3
```

```python
df4 = pd.DataFrame({'ASCII_Numbers': np.arange(71,75),
                    'characters': [chr(i) for i in range(71, 75)]})

df4
```

```python
merged_df = pd.merge(df1, df2, on='ASCII_Num', how='outer')
merged_df
```

```python
merged_df = pd.merge(df1, df2, on='ASCII_Num', how='inner')
merged_df
```

```python
merged_df = pd.merge(df1, df2, on=['ASCII_Num', 'char'], how='outer')
merged_df
```

```python
merged_df = pd.merge(df1, df2, on=['ASCII_Num', 'char'], how='inner')
merged_df
```

```python
merged_df = df1.merge(df3, on=['ASCII_Num'], how='outer')
merged_df
```

```python
merged_df = df1.merge(df3, on=['ASCII_Num'], how='inner')
merged_df
```

```python
merged_df = df1.merge(df3, on=['ASCII_Num'], how='left')
merged_df
```

```python
merged_df = df1.merge(df3, on=['ASCII_Num'], how='right')
merged_df
```

## Drop_duplicates()
- for dropping the dupicate values

```python
df = pd.DataFrame({'A': [1, 2, 1, 2],
                   'B': [100, 200, 100, 400],
                   'C':[4, 4, 5, 5]})
df
```

```python
df.drop_duplicates(['C'])
```

```python
df.drop_duplicates(['A'], keep='first')
```

```python
df.drop_duplicates(['A'], keep='last')
```

```python
df.drop_duplicates(['A', 'B'], keep='first')
```

```python
df.drop_duplicates(['A', 'B'], keep='last')
```

## Duplicated

```python
df = pd.DataFrame({'A': [1, 2, 1, 2],
                   'B': [100, 200, 100, 400],
                   'C':[4, 4, 4, 5]})
df
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
df[~(df.duplicated())].reset_index(drop=True)
```
