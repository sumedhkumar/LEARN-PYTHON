### Creating A DataFrame

```python
data = {"column A": [2, 3, 4, 5], "column B": [9, 8, 7, 6]}
type(data)
```

```python
dataframe = pd.DataFrame(data)
dataframe
```

```python
# Columns in DataFrame --> Features = Variables = Fields = Dimensions
# Rows in DataFrame --> Records = Values = Observations = Index
```

```python
data2 = np.array([33, 44, 55, 66])
data2
```

```python
df = pd.DataFrame(data2, columns=['values'])
df
```

```python
data3 = [44, 55, 66, 77]

df = pd.DataFrame(data3, columns=['Vals'])
df
```

### How to add a column

```python
dataframe = pd.DataFrame(data)
dataframe
```

```python
dataframe['column C'] = "AI"
dataframe
```

```python
dataframe['column C'] = "GenAI" # Overwrite
dataframe
```

```python
dataframe['column C'] = ['G', 'E', 'N', "AI"]
dataframe
```

```python
dataframe['column C'] = list('gene')
dataframe
```

```python
dataframe['column D'] = np.array([100, 200, 300, 400])
dataframe
```

```python
# Add a column E

dataframe['column D'] = [1, 2, 3, 4]
dataframe
```

```python
dataframe['column E']=['genz','gena','genb', 'genG']
dataframe
```

## Setting a column as index

```python
dataframe['index'] = np.array([1, 2, 3, 4])
dataframe
```

```python
df = dataframe.set_index('index')
df
```
