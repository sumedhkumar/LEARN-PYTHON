### Groupby & Get_group

```python
df.groupby('genre')
```

```python
df.groupby('genre').first()
```

```python
df.groupby('genre').last()
```

```python
df
```

```python
df.groupby('genre').get_group('Adventure')
```

```python
df.groupby('content_rating').get_group('PG-13')
```

```python
df.groupby('star_rating').get_group(9.0)
```

```python
df.groupby(['content_rating', 'genre']).get_group(('R', 'Crime'))
```

```python
df.groupby(['content_rating', 'genre'])['duration'].sum()
```

```python
df.groupby(['content_rating', 'genre'])['duration'].sum().reset_index()
```

## LOC

```python
df.loc[df['genre']=='Crime']
```

```python
df.loc[(df['genre']=='Crime')&(df['content_rating']=='PG')] # It will consider both the conditions mentioned
```

```python
df.loc[(df['genre']=='Crime')|(df['content_rating']=='PG')] # Or will consider either of one condition needs to be true
```

```python
df.loc[(df['star_rating']<9.0)&(df['star_rating']>8)]
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/movies_data.csv")
df
```

```python
df.loc[(df['content_rating']=='R')&(df['star_rating']>=9.0)]
```

```python
df.loc[~((df['star_rating']<=9.0)&(df['star_rating']>=8))] # ~ not equals to
```

```python
df.loc[(df['genre']=="Action")|(df['star_rating']<7)]
```
