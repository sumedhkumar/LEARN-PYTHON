## Ways to print any statement in Python 

```python
a = (1, 9, 0)
a = 66   # >>> entities which stores any value are called variables
b = 44

print(a, b)
```

```python
print("value of a = ", a , "\nValue of b = ", b)
```

```python
print("Value of a and b are ", a, "&", b)
```

```python
# %i , %d , %f , %s
print("Value of a = %i and b = %d" % (a, b))
```

```python
# . format method
print("Value of a = {} and b = {}".format(a, b))
```

```python
# f-string
print(f"Value of a = {a} and b = {b}")

sports = "Cricket"

print(f"I want to play {sports}")
```

```python
# FR strings

sports = "Chess" 
# r is working as a escape charecter (r is used to print the raw string) (\)
print(r"I want to play {sports} \tlike samay raina")
print("---------------------------------")
print(f"I want to play {sports} \tlike samay raina")
print("---------------------------------")
print(fr"I want to play {sports} \tlike samay raina")
```

## Taking in Variable using INPUT Function

```python
# Input function in Python
```

```python
a = 44
b = 33

print(f"Value of a and b are {a} and {b}")
```

```python
a = input("Enter the value of a: ")
b = input("Enter the value of b: ")

print(f"Value of a and b are {a} and {b}, type of a is {type(a)} and type of b is {type(b)}, a + b = {a + b}")
```

```python
a = int(input("Enter the value of a: ")) # >>> input function always takes the value in string format then we converted it to integer
b = int(input("Enter the value of b: "))

print(f"Value of a and b are {a} and {b}, type of a is {type(a)} and type of b is {type(b)}, a + b = {a + b}")
```

```python
a = '66'
type(a)

a = int(a)
type(a)
```

```python
a = float(input("Enter the value of a: ")) # >>> input function always takes the value in string format then we converted it to float
b = float(input("Enter the value of b: "))

print(f"Value of a and b are {a} and {b}, type of a is {type(a)} and type of b is {type(b)}, a + b = {a + b}")
```
