Linear Regression:
	- Linear Regression is a machine learning algorithm that falls under supervised learning.
	- Linear regression is a basic and commonly used type of predictive analysis.
    - The model basically describe the relationship between the targeted column & the input columns by fitting a line on the observed data.
    - the target variable is linearly related to the input variable.
    - It is also called as ordinary least squared algorithm.
	- Linear regression is a linear model, e.g. a model that assumes a linear relationship between the input variables (x) and the single output variable (y). More specifically, that y can be calculated from a linear combination of the input variables (x).

# How does a best fit line is found by machine?

```python
		- The best fit line in linear regression model is obtained by eq. y=mx+c where y is dependent variable, x is independent variable, m is slope of the best fit line & c is the intercept on y axis.
		- The best fit line always passes through he mean point.
Ones we fit the model into linear regression then in blackened it uses gradient decent algorithm which in itself uses bond driver algorithm using which it finds global minima point which gives up exact values of m.
```
