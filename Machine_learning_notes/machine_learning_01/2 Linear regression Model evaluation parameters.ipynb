# Adjusted R^2

```python
R2 = 0.85
Adj R2 = 0.85

Case 1 : Added Good Predictor (R > 0.7 or R < -0.7)  >> correlated feature
        R2 : 0.87
        Adj R2: 0.87
            
Case 1 : Added Bad Predictor (R < 0.7 or R > -0.7)  >> Non-correlated feature
        R2 : 0.86
        Adj R2: 0.84
            

The value of adjusted R2 increases only for Good Predictor. For Bad predictor the value decreases.

Adj R2 = 1 - ((1-R2)*(N-1)/(N-P-1))
```

# Variance

```python
Variance is the spread of the total data points on the graph.

If the spread is high (the data points are spread all over the graph) then we say Variance is high.
If the spread is low (the data points are closely spread over the graph) then we say Variance is low.

If the variance increases then the accuracy of the model would decrease & voice versa.

Var = sum(Xi-Xm)**2 / N

where, 
    sum = summation
    Xi = Actual Points
    Xm = Mean of all the Actual points 
    N = No. of points (population)
    
```

# Standard Deviation

```python
Standard Deviation (STD) :
    It explains how far the data points are from the mean point.
    It is simply the sqrt of Variance
    It also explains how much proportion of the data is how many std away.
    
Variance is the spread of the total data points on the graph.

If the spread is high (the data points are spread all over the graph) then we say Variance is high.
If the spread is low (the data points are closely spread over the graph) then we say Variance is low.

If the variance increases then the accuracy of the model would decrease & voice versa.

Var = sqrt(sum(Xi-Xm)**2 / N)

where, 
    sum = summation
    sqrt = square root
    Xi = Actual Points
    Xm = Mean of all the Actual points 
    N = No. of points (population)
    
```
