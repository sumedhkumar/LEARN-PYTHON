### Indexing in List

```python
# Indexing in List :
    # Positive indexing : It starts with 0 (zero) & goes to +ve infinity (left to right)
    # Negative Indexing: It starts with -1 & goes to -ve infinity (Right to left)
```

```python
list1 = [1, 5, 7, "Sehwag"]
# Left -------------------> Right ---> Positive indexing 
# Left <------------------- Right ---> Negative indexing 
list1[3]
```

```python
list1[-1]
```
