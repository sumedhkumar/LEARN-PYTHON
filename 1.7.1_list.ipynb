## List

```python
# data type --> List
# List is the collections of items, entities, that are seperated by (,) comma & this should be inside square braces []
# List is use to store various data types ---> Int, Strings, Floats, set, dict, dataframes, another list init etc
# List allows duplicated items 
# List is Mutable == changeable == editable data type

#eg. 
new_list = [1, 0.2, 3, 3, "strings", (1, 2, 3), {"name": "Shahiba"}, [1, 2, 4, 4]]   #<---- square braces
new_list
```

### Indexing in List

```python
# Indexing in List :
    # Positive Indexing : It starts with 0 (zero) & goes to +ve infinity (left to right)
    # Negative Indexing: It starts with -1 & goes to -ve infinity (Right to left)
```

```python
list1 = [1, 5, 7, "Sehwag"]
# Left -------------------> Right ---> Positive indexing 
# Left <------------------- Right ---> Negative indexing 
list1[3]
```

```python
list1[-1]
```

```python
#Q : How to get "7"

list1[2], list1[-2]
```

## Slicing

```python

# slicing = list[start index : end index : step size]  <---- Syntax 
# start index default = 0 
# end index default = len(list)
# step size default = 1


list1 = [1, 4, 7, "Virat", 8, 1]
list1[0 : len(list1) : 1] # <---slicing
```

```python
len(list1)
```

```python
list1[0 : 3]
```

```python
list1[3 : 6]
list1[3: len(list1)]
list1[-3:]
```

```python
list1 = [1, 4, 7, 'Virat', 8, 1]  # Mr Mario game 

list1[0: len(list1): 2]
```

```python
# reverse the given list

list1 = [1, 4, 7, 'Virat', 8, 1] 
        #-6 -5 -4   -3     -2 -1  ---> negative index

list1[::-1] # step size -1 
```

```python
list1[::-2]
```

```python
list1[-5:-1]
```

```python
list1[-1:-5:-1]
```

```python
# Loop functions we access list items
    # 1. For loop 
    # 2. While loop
    
    
cities = ["Sambhajinagar", "Pune", "Mumbai", "dhule", "Nagpur"]

for i in cities: 
    print(i)
```
