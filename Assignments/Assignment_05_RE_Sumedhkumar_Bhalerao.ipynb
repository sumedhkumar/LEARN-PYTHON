Q 11. Write a Python program to find sequences of lowercase letters joined with a underscore

```python
string = "aab_cbbbc aab_Abbbc Aaab_abbbc bec__ome today defi"

X = re.findall(r'\b[a-z]+[_]+[a-z]{1,19}\b',string)
X
```

```python
def text_match(text):
        patterns = '^[a-z]+_[a-z]+$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aab_cbbbc"))
print(text_match("aab_Abbbc"))
print(text_match("Aaab_abbbc"))
```

Q 15. Write a program that matches a word at the end of string, with optional punctuation.

```python
def text_match(string):
        patterns = '.+\W$'
        if re.findall(patterns, string):
            return 'Found a match!'
        else:
            return('Not matched!')

print(text_match("""Elon Reeve Musk FRS is an entrepreneur, investor, and business magnate.
He is the founder, CEO, and Chief Engineer at SpaceX; early-stage investor, CEO, and
Product Architect of Tesla, Inc."""))

print(text_match("""Elon Reeve Musk FRS is an entrepreneur, investor, and business magnate.
He is the founder, CEO, and Chief Engineer at SpaceX;"""))
```

Q 16. Write a Python program that matches a word containing 'z'.
