```python
# Lambda
# Lambda function : Anonymous function
# a function without name 
# it is defined as lambda keyward 
# Here, we do not need to return or yeild 
# Short hand function
```

```python
def square(n): # defined a name of a function
    return n**2

square(10)
```

```python
# Syntax of lambda
# lambda Arguments : expression (operation on the Arguments)
# ex: 
lambda a : a**2, 
lambda a, b: a + b 
```

```python
sqr = lambda n: n**2 #short hand expression

sqr(10)
```

```python
(lambda n: n**2)(9)
```

```python
def sum(a, b):
    return a + b

sum(4, 5)
```

```python
sum = lambda a, b: a+b

sum(4, 5)
```

```python
def add(a, b, c):
    return (a+b) * c

add(3, 4, 5)
```

```python
add = lambda a,b,c : (a+b) *c
add(3,4,5)
```

```python
add = lambda a, b, c=10: (a+b) *c  # here c is default arg
add(3,4)
```

```python
add = lambda a, b, c=10: (a+b) *c  # here c is default arg
add(3,4,3)
```

```python
a = 2
b = 3
c = 10

x = st.mean((a, b, c))
x
```

```python
x = lambda a,b,c : st.mean((a,b,c))

x(2, 3, 4)
```

```python
def get_mean(a, b, c):
    
    y = st.mean((a, b, c))
    
    return (round(y, 4))

get_mean(2, 3, 3)
```
