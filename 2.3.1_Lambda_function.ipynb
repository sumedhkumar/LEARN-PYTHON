```python
# Syntax of lambda
# lambda Arguments : expression (operation on the Arguments)
# ex: 
lambda a : a**2,
lambda a, b: a + b 
```

```python
def multiplication(x,y):
    return x * y
# mul(6, 9)

x = lambda a, b: multiplication(a, b) - (a + b)

x(2, 3)
```

```python
def multiplication(x,y):
    return x * y
# mul(6, 9)

x = lambda a, b, c: [(multiplication(a, b) - (a + b))* i for i in c]

x(2, 3, [1, 2, 3, 4])
```
