## If else statements

```python
# Defination of if and else statements

# If else statements are decision makings statements (Check if a condition is true or false) 
# In case when we have only if and else:
# If the condition is true, the if block will be executed and
# If the condition is false, the else block will be executed


# We dont write any condition in the else block, it will be executed if the condition in the if block is false
# If and elif should have a condition, else should not have a condition

# If & elif statement can be used alone without else statement; no need to have else statement

# Valid statements: 
# If statement
# If and else statement
# If elif and else statement
# If and elif statement
# If elif and elif... elif statement

# Invalid statements:
# If else statement without if statement
# If else elif (wrong order)
```

```python
raw = "Python is very much easy"

if "Python" in raw: # Check if Python is in the raw ----> True
    a = 5 + 100
    print(a)
    print("Yes, Python is in the raw")
else:
    print("No, Python is not in the raw")
```

```python
# Question: 
#  less than 30% --> Fail 
#     30% to 60% --> Average
#     60% to 80% --> Good
#     80% to 100% --> Excellent

score = int(input("Enter your score: "))
if score < 30: 
    print("Fail")
elif score >= 30 and score < 60:
    print("Average")
elif score >= 60 and score < 80:
    print("Good")
elif score >= 80 and score <= 100:
    print("Excellent")
else:
    print("Invalid score")
```

### Nested If else statements

```python
# Nested if else statement

# Within one if block, we can have another if block
# If, elif and else block in another if else block

# Nested if else statement is used to check multiple conditions
```

```python
score = 65

if score < 80: # False
    if score < 60: # False
        if score < 30: # False
            print("Fail")
        else:
            print("Average")
    else:
        print("Good")
else:
    print("Excellent")
```
