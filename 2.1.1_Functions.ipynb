- Yeild

```python
def even_num(number):
    if number % 2 == 0:
        yield "even"
    else:
        yield "odd"
    
X = even_num(112)
X
```

```python
next(X)
```

```python
list(X)
```

```python
def max_num(a):
    yield a

list1 = [1, 3, 4]
for i in list1:
    op = max_num((i))
```
