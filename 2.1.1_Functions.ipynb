```python
def get_smallest_number(a, b):
    global z # global variable
    z = 3 
    if a < b and a < z:
        return a
    elif b < a and b < z:
        return b
    else:
        return z
    
a = 10 # global variable
b = 20

get_smallest_number(a, b)
```
